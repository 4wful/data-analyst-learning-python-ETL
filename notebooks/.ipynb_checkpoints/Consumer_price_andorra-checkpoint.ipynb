```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cargar el dataset saltar la fila 1
df = pd.read_csv("consumer-price-indices_and.csv", skiprows=1)
#verificar su carga
df.head()
```

LIMPIEZA DE DATOS

```python
#Primero cambiaremos los nombres de las columnas
df.columns = [
    "country_code", "date_start", "date_end", "unknown_1", "unknown_2", 
    "country_name", "indicator_code", "indicator_name", "unknown_3", 
    "unknown_4", "unknown_5","month","unknown_6", "year", "indicator_type", "indicator_value", "unknown_7","unknown_8"
]
#verificar cambios
df.head()
```

```python
#Ahora eliminare las columnas innecesarias 
df = df.drop(columns=["unknown_1", "unknown_2", "unknown_3", "unknown_4", "unknown_5", "unknown_6","indicator_type","unknown_7","unknown_8"])
#verificaré cambios
df.head()
```

CONVERSIÓN DE DATOS

```python
# Convertir fechas a tipo datetime
df["date_start"] = pd.to_datetime(df["date_start"])
df["date_end"] = pd.to_datetime(df["date_end"])

# Convertir el año a tipo entero
df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")

# Convertir el valor del indicador a tipo numérico
df["indicator_value"] = pd.to_numeric(df["indicator_value"], errors="coerce")
```

```python
#Verificar Valores nulos
print(df.isnull().sum())
```

```python
#Como no hay valores nulos procedemos a guardar el archivo limpio y la filtración de datos 
df.to_csv("consumer_price_and_clean.csv", index=False)
```

EJERCICOS
- Filtra solo los años entre 2010 y 2020.

- Calcula la media, mediana y desviación estándar del índice (Value) en ese período.

- Agrupa los datos por año y calcula el promedio anual.

- Haz un gráfico de barras con la evolución del índice promedio por año.

```python
#Filtra solo los años entre 2010 y 2020.
df_2010_2020 = df[(df["year"] >= 2010) & (df["year"] <= 2020)]
#verificar el nuevo df
df_2010_2020.head()
```

```python
#Calcula la media, mediana y desviación estándar del índice (Value) en ese período.
#Calcular media del periodo 2010-2020
media = np.mean(df_2010_2020["indicator_value"])
print(f"La media del valor entre el periodo 2010-2020 es: {media}")
#Calcular mediana del periodo 2010-2020
mediana = np.median(df_2010_2020["indicator_value"])
print(f"La mediana del valor entre el periodo 2010-2020 es: {mediana}")
#Calcular desviación estándar del periodo 2010-2020
des_estandar = np.std(df_2010_2020["indicator_value"])
print(f"La desviación estándar del valor entre el periodo 2010-2020 es: {des_estandar}")
```

```python
#Agrupa los datos por año y calcula el promedio anual.
# Agrupar los datos por año y calcular el promedio anual
promedio_anual = df_2010_2020.groupby("year")["indicator_value"].mean()
# Mostrar el resultado
print(f"El Promedio anual por año es {promedio_anual}")
```

```python
# Graficar la evolución del índice promedio por año (gráfico de barras)
plt.figure(figsize=(10, 6))
plt.bar(promedio_anual.index, promedio_anual.values, color='skyblue')
plt.title('Evolución del índice promedio por año')
plt.xlabel('Año')
plt.ylabel('Promedio')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
